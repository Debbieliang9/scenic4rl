```python
from matplotlib import pyplot
pyplot.figure(figsize=(24, 6))

hue_order = ['PPO with No Pretraining', 'PPO with Pretraining', 'BC Policy', 'Scenic Policy'] 
sns.set(font_scale=1.5)
ax = sns.barplot(x="Scenario", y="Average Goal Difference", hue='Policy Type',  hue_order = hue_order, data=df)

from matplotlib.ticker import MultipleLocator
ax.yaxis.set_major_locator(MultipleLocator(0.2))
ax.set(xlabel=None)

ax.figure.savefig("pretraining.png", bbox_inches="tight")
```
